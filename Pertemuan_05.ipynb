```python
batas_atas = 500

# Perulangan 'for' yang menggunakan fungsi range(1, 501).
# Fungsi range(start, stop) akan menghasilkan angka dari 'start'
# (inklusif/termasuk) hingga 'stop' (eksklusif/tidak termasuk).
for angka in range(1, batas_atas + 1):
    print(angka)
```

```python
# Program Python untuk Menampilkan Bilangan Ganjil (1 - 200) menggunakan 'while'

# 1. Inisialisasi variabel penghitung.
# Kita mulai dari 1 karena 1 adalah bilangan ganjil pertama.
angka = 1

# 2. Kondisi Perulangan 'while'.
# Perulangan akan terus berjalan selama nilai 'angka' kurang dari atau sama dengan 200.
while angka <= 200:

    # 3. Badan Perulangan
    # Tampilkan nilai 'angka' saat ini
    print(angka)

    # 4. Perubahan (Update/Iteration)
    # Tambahkan 2 ke 'angka' agar langsung lompat ke bilangan ganjil berikutnya.
    # Contoh: 1 + 2 = 3, 3 + 2 = 5, dst.
    angka += 2

# Perulangan akan berhenti saat 'angka' menjadi 201
```

```python
# Data Penjualan Harian
penjualan_harian = [450, 252, 115, 75, 20]
stok_awal = 1000

# =================================================================
# Bagian A: Menghitung Sisa Stok Akhir (Menggunakan FOR loop)
# =================================================================
print("=== Bagian A: Sisa Stok Akhir (FOR loop) ===")
sisa_stok = stok_awal

# Menggunakan for loop untuk mengurangi stok dengan total penjualan
for terjual in penjualan_harian:
    sisa_stok -= terjual # Operasi Pengurangan

print(f"\nSisa stok produk X setelah 5 hari adalah: {sisa_stok} unit.")

# =================================================================
# Bagian B: Kapan Stok Menipis (Menggunakan WHILE loop)
# =================================================================
print("\n=== Bagian B: Kapan Stok Turun di Bawah Batas (WHILE loop) ===")
batas_minimum = 450
stok_saat_ini = stok_awal
hari_ke = 0 # Indeks untuk list penjualan_harian

# Menggunakan while loop untuk melanjutkan selama stok masih di atas atau sama dengan batas
# DAN masih ada data penjualan yang tersisa
while stok_saat_ini >= batas_minimum and hari_ke < len(penjualan_harian):

    # Ambil jumlah terjual untuk hari berikutnya
    terjual_hari_ini = penjualan_harian[hari_ke]

    # Lakukan Pengurangan Stok
    stok_saat_ini -= terjual_hari_ini

    # Pindah ke hari berikutnya (Hari ke-1, Hari ke-2, dst.)
    hari_ke += 1

    print(f"Akhir Hari ke-{hari_ke}: Sisa stok = {stok_saat_ini} unit")

# Setelah loop selesai
if stok_saat_ini < batas_minimum:
    print(f"\nStok TURUN di bawah batas minimum {batas_minimum} unit pada akhir hari ke-{hari_ke}.")
else:
    print(f"\nStok tidak pernah turun di bawah batas minimum {batas_minimum} unit dalam 5 hari.")
```
